```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

# Total revenue by products

```python
total_rev=df.groupby("Product")["Revenue"].sum()
total_rev
```

# Average units sold per product

```python
avg_units=df.groupby("Product")["Units_Sold"].mean()
avg_units
```

# Line Chart

```python
revenue_trend=df.pivot(index="Month", columns="Product", values="Revenue")
revenue_trend.plot(kind="line")
plt.title("Revenue trend per product")
plt.xlabel("Months")
plt.ylabel("Revenue")
#plt.xticks(rotation=45)
```

# Bar chart

```python
total_rev.plot(kind="bar")
plt.title("Revenue per product")
plt.xlabel("Products")
plt.xticks(rotation=45)
plt.ylabel("Revenue")
```

```python
total_rev.plot(kind="pie")
plt.title("Pie chart on Revenue per product")
```
